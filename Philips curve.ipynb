```python
# phillips_curve.py
# A simple economics project: Inflation vs Unemployment (Phillips Curve)

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats

# Sample economic data (you can later replace with real data)
data = {
    "Year": [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    "Inflation": [5.8, 4.9, 3.6, 3.9, 4.5, 6.2, 5.1, 6.8, 5.9],  # in %
    "Unemployment": [6.2, 5.8, 5.5, 5.3, 5.1, 8.0, 7.2, 6.5, 6.0]  # in %
}

df = pd.DataFrame(data)

# Correlation between inflation and unemployment
corr = df["Inflation"].corr(df["Unemployment"])
print(f"Correlation between Inflation and Unemployment: {corr:.2f}")

# Simple regression (Phillips curve idea)
slope, intercept, r, p, std_err = stats.linregress(df["Unemployment"], df["Inflation"])

# Function for regression line
def predict(x):
    return slope * x + intercept

# Plot
plt.scatter(df["Unemployment"], df["Inflation"], color='blue', label='Data Points')
plt.plot(df["Unemployment"], predict(df["Unemployment"]), color='red', label='Regression Line')
plt.title("Inflation vs Unemployment (Phillips Curve)")
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Inflation Rate (%)")
plt.legend()
plt.grid(True)
plt.show()
```
